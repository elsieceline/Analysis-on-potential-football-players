Data Cleaning and Preprocessing

Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix

```python
## function for converting M and K with their respective values and for replacing the euro symbol for data manupliation
def convert_value(val):
    if isinstance(val, str):
        if val.endswith('M'):
            return float(val.replace('€', ' ').replace('M', ' '))*1000000
        elif val.endswith('K'):
            return float(val.replace('€', ' ').replace('K', ' '))*1000
        else:
            return float(val.replace('€', ''))
    else:
        return val

        
```

```python
df['Value'] = df['Value'].apply(convert_value)
df['Wage'] =  df['Wage'].apply(convert_value)
df['Release Clause'] =  df['Release Clause'].apply(convert_value)
```

```python
df[['Value', 'Wage', 'Release Clause']].head()
```

Convert the column "Joined" into integer data type with keeping only the year

```python
## Converting Joined column to datetime format for easy extraction of year
df['Joined'] = pd.to_datetime(df['Joined'])
```

```python
df['Joined'] = df['Joined'].dt.strftime('%Y')
```

```python
## Using errors = 'coerce' will fill any non numeric columns with nan and fillna(0) to fill that nan with 0
df['Joined'] = pd.to_numeric(df['Joined'], errors = 'coerce').fillna(0).astype(int)
```

```python
df['Joined'].head()
```

Convert the column "Contract Valid Until" to pandas datetime type

```python
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'], errors='coerce')
```

```python
df['Contract Valid Until'].head()
```

```python
df['Contract Valid Until'] = df['Contract Valid Until'].dt.strftime('%Y')
```

```python
df['Contract Valid Until'].head()
```

The column 'Height' is in inches with a quotation mark, Convert to float with decimal points

```python
## Replacing single and double quotation mark using str.replace() method
## Removing any leading or trailing whitespaces using str.strip()
df['Height'] = df['Height'].str.replace("'","", regex = False).str.replace(' " ', '', regex = False).str.strip().astype(float)
```

```python
df['Height']
```

The column "Weight" has the suffix as lbs, remove the suffix and convert to float

```python
df['Weight'] = df['Weight'].str.replace("lbs", "", regex = False).astype(float)
```

```python
df['Weight']
```

Check for the percentage of missing values and impute them with appropriate imputation techniques

```python
(df.isna().sum()/len(df))*100
```

```python
## Replacing missing values with mean for numerical columns
for col in numerical_columns:
   df[col].fillna(df[col].mean(), inplace = True)
## Replacing missing values with mode for categorical columns
for col in categorical_columns:
    df[col].fillna(df[col].mode(), inplace = True)
```

Generate a data frame containing top 5 players by Overall rating for each unique position

```python
grouped_by_position = df.groupby('Position')

top_5_players_by_position = []

for position, group in grouped_by_position:
    top_5_players = group.nlargest(5, 'Overall')
    top_5_players_by_position.append(top_5_players)

top_5_players_df = pd.concat(top_5_players_by_position)

top_5_players_df.reset_index(drop=True, inplace=True)

top_5_players_df
```

What is the average wage one can expect to pay for the top 5 players in every position

```python
grouped_by_position = df.groupby('Position')
average_wage_top_5_players_by_position = grouped_by_position.apply(lambda x: x.nlargest(5, 'Overall')['Wage'].mean())
print("Average wage for the top 5 players in every position:")
print(average_wage_top_5_players_by_position)
```
